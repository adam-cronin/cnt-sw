# Rotating two bonded atoms about their midpoint 

```python
#input an xyz file below and define b1 and b2 to be coords of rotating atoms and choose a third point to define the plane (midpoint of b3 and b4)
#the code will create new input and xyz files for the new coordinates
#be sure a folder is created with the same name as below
import os
cnt_file=os.path.join('nt_55_20opt.xyz') #adjust input file here
name='nt5520a1ts' #this will be the name of the input and xyz files (make sure folder is created with same name)
import numpy
cnt_coords=numpy.genfromtxt(fname=cnt_file,dtype='unicode',skip_header=2)
coords=(cnt_coords[:,1:])
coords=coords.astype(numpy.float)
num=117
b1=coords[num]                 
b2=coords[num+1]
b3=coords[num+20]   #will need to be adjusted if nanotube size changes
b4=coords[num+21]
x1=b2[0]
x2=b1[0]
y1=b2[1]
y2=b1[1]
z1=b2[2]
z2=b1[2]
x3=(b3[0]+b4[0])/2
y3=(b3[1]+b4[1])/2
z3=(b3[2]+b4[2])/2
r=0.5*numpy.sqrt((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)
M=[(b1[0]+b2[0])/2,(b1[1]+b2[1])/2,(b1[2]+b2[2])/2]
def equation_plane(x1, y1, z1, x2, y2, z2, x3, y3, z3):  
    a1 = x2 - x1 
    b1 = y2 - y1 
    c1 = z2 - z1 
    a2 = x3 - x1 
    b2 = y3 - y1 
    c2 = z3 - z1 
    a = b1 * c2 - b2 * c1 
    b = a2 * c1 - a1 * c2 
    c = a1 * b2 - b1 * a2 
    d = (- a * x1 - b * y1 - c * z1) 
    return a,b,c,d
s=equation_plane(x1,y1,z1,x2,y2,z2,x3,y3,z3)
a=s[0]
b=s[1]
c=s[2]
d=s[3]

from scipy.optimize import fsolve
import numpy
def xyz_eq(v):
    x=v[0]
    y=v[1]
    z=v[2]
    theta=55*numpy.pi/180     #angle of rotation
    F=numpy.empty((3))
    F[0]=(x-M[0])**2+(y-M[1])**2+(z-M[2])**2-r**2
    F[1]=numpy.sqrt((x-x1)**2+(y-y1)**2+(z-z1)**2)-2*r*numpy.sin(theta/2)
    F[2]=a*x+b*y+c*z+d
    return F
zguess=numpy.array([x1,y1,z1])
j=fsolve(xyz_eq,zguess)

x=j[0]-2*(j[0]-M[0])
y=j[1]-2*(j[1]-M[1])
z=j[2]-2*(j[2]-M[2])

newcoords=(f"""C {x} {y} {z}
C {j[0]} {j[1]} {j[2]}""")
old_str=f"""C {x2} {y2} {z2}
C {x1} {y1} {z1}"""
with open(cnt_file) as f:     
    xyz=data=f.read()

newxyz=xyz.replace(f"{x2}",f"{x}").replace(f"{y2}",f"{y}").replace(f"{z2}",f"{z}").replace(f"{x1}",f"{j[0]}").replace(f"{y1}",f"{j[1]}").replace(f"[z1]",f"j[2]")

with open(os.path.join(f'{name}',f'{name}.xyz'),'w+') as f:
    f.write(newxyz)
    
file=open(cnt_file,'r')
data=file.readlines()    
no_header=newxyz.replace(data[0],"").replace(data[1],"")

f=open(os.path.join(f"{name}",f"{name}.in"),"w+")
f.write("""molecule{
""")
f.write(no_header)
f.write("""}
memory 20 GB
set basis 6-31G(d)
set reference rhf   #these will need to be adjusted depending on the experiment 
set opt_type min
optimize('pbe')""")
f.close()
```
