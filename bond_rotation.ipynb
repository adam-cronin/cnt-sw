```python
import os
cnt_file=os.path.join('nt_5_5_5_opt.xyz')
import numpy
cnt_coords=numpy.genfromtxt(fname=cnt_file,dtype='unicode',skip_header=2)
coords=(cnt_coords[:,1:])
coords=coords.astype(numpy.float)
```

```python
b1=coords[28]
b2=coords[29]
print(b1,b2)
```

```python
import sympy as sym

x1=b1[0]
x2=b2[0]
y1=b1[1]
y2=b2[1]
z1=b1[2]
z2=b2[2]
r=0.5*numpy.sqrt((x2-x1)**2+(y2-y1)**2)
dy=(y1-y2)/2
print(dy)
dx=(x1-x2+2*r)/2
print(dx)
import math 
beta=math.asin(dy/r)
print(r)
print(beta)
```

```python
z_vals=b1[2]
z_vals_=b2[2]
print(z_vals,z_vals_)
```

```python
theta=numpy.pi/4
import numpy
def z_eqn(theta):
    z=z_vals+2*r*numpy.sin(0.5*theta)*numpy.cos(0.5*theta)
    return z
new_z=z_eqn(theta)
def z_eqn_(theta):
    z=z_vals_-2*r*numpy.sin(0.5*theta)*numpy.cos(0.5*theta)
    return z
new_z_=z_eqn_(theta)
print(new_z,new_z_)
def line_eq(x1,x2,y1,y2):
    m=(y2-y1)/(x2-x1)
    b=m*(-x2)+y2
    return m,b
m,b=line_eq(x1,x2,y1,y2)
```

```python
from scipy.optimize import fsolve
def distance_xy(theta):
    d=2*r*numpy.sin(0.5*theta)**2
    return d
d=distance_xy(-theta)
d2=distance_xy(-theta-numpy.pi)
def xy_eq_(z):
    x=z[0]
    y=z[1]

    F=numpy.empty((2))
    F[0]=numpy.sqrt((x-x2)**2+(y-y2)**2)-d
    F[1]=y-m*x-b
    return F
zguess=numpy.array([x2,y2])
z=fsolve(xy_eq_,zguess)
print(z)
x_2=z[0]
y_2=z[1]
def xy_eq(z):
    x=z[0]
    y=z[1]

    F=numpy.empty((2))
    F[0]=numpy.sqrt((x-x2)**2+(y-y2)**2)-d2
    F[1]=y-m*x-b
    return F
zguess=numpy.array([x2,y2])
z=fsolve(xy_eq,zguess)
print(z)
x_1=z[0]
y_1=z[1]
```
